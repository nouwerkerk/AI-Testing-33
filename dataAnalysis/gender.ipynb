**Data Analysis**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Let's load the dataset
data = pd.read_csv('../data/all_data.csv')

# Let's specify the features and the target
y = data['checked']
X = data.drop(['checked'], axis=1)
X = X.astype(np.float32)
```

```python
for i in (list(data.columns.values)):
    print(i)
```

```python
geslacht_vrouw = data['persoon_geslacht_vrouw']
counts = geslacht_vrouw.value_counts()
print(counts)

plt.barh([0, 1], counts.values, color=['blue', 'orange'])

# Add labels and title
plt.xlabel('Count')
plt.title('Men/Women Count')

plt.yticks([0, 1], labels=['men', 'women'])
plt.plot
```

```python
relative_freq = counts / np.sum(counts)
real_freq = np.array([338343, 332267]) / (338343+332267)
print(real_freq)

y_pos = np.arange(2)

plt.barh(y_pos, relative_freq.values, height=0.4, label='Dataset', color=['orange'])

plt.barh(y_pos - 0.4, real_freq, height=0.4, label='Real', color=['blue'])

# Add labels and title
plt.xlabel('Relative occurence (%)')
plt.title('Men/Women Relative Occurence')

plt.legend()

plt.yticks([-0.2, 0.8], labels=['men', 'women'])

plt.tight_layout()
plt.show()
```
