```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../data/all_data.csv')
```

#### General information

```python
taaleis_voldaan = data['persoonlijke_eigenschappen_taaleis_voldaan']
counts = taaleis_voldaan.value_counts()

plt.barh(counts.index, counts.values, color=['blue', 'orange', 'grey'])

# Add labels and title
plt.xlabel('Count')
plt.title('Language requirement met')

plt.yticks([0, 1, 2], labels=['not_met', 'met', 'unknown'])
plt.show()
```

```python
sinds_taaleis = data['persoonlijke_eigenschappen_dagen_sinds_taaleis']

print(len(sinds_taaleis[0 <= sinds_taaleis]))

counts = sinds_taaleis.value_counts()

plt.bar(counts.index, counts.values)

# Add labels and title
plt.ylabel('Count')
plt.title('Days since language requirement met')

plt.show()

plt.show()
```

```python
taaleis_voldaan = taaleis_voldaan[taaleis_voldaan < 2]

# print(taaleis_voldaan)

relative_freq = np.array(taaleis_voldaan.value_counts(normalize=True).sort_index())
print(relative_freq)

real_freq = np.array([1-0.4353, 0.4353])
# print(real_freq)
```

```python
y_pos = np.arange(2)

plt.barh(y_pos, relative_freq, height=0.4, label='Dataset', color='orange')

plt.barh(y_pos - 0.4, real_freq, height=0.4, label='Real', color='blue')

plt.yticks([y_pos[0]-0.4, y_pos[0], y_pos[1]-0.4, y_pos[1]],
           np.array(["Non native","Language requirement not met","Native","Language requirement met"]))

plt.xlabel('Relative occurance (%)')
plt.title('Language Requirement Met vs (Non) Native Inhabitants of Rotterdam')

plt.legend()

# Show the plot
plt.tight_layout()
plt.show()

```
