```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../data/all_data.csv')
```

#### General information

```python
taaleis_voldaan = data['persoonlijke_eigenschappen_taaleis_voldaan']
counts = taaleis_voldaan.value_counts()

plt.bar(counts.index, counts.values, color=['blue', 'orange', 'grey'])

# Add labels and title
plt.ylabel('Count')
plt.title('Language requirement met')

plt.xticks([0, 1, 2], labels=['not_met', 'met', 'unknown'])
plt.show()
```

```python
taaleis_voldaan = taaleis_voldaan[taaleis_voldaan < 2]

relative_freq = np.array(taaleis_voldaan.value_counts(normalize=True).sort_index())
print(relative_freq)

real_freq = np.array([1-0.4353, 0.4353])
print(real_freq)
```

```python
y_pos = np.arange(2)

plt.barh(y_pos, relative_freq, height=0.4, label='Dataset distr.', color='orange')

plt.barh(y_pos - 0.4, real_freq, height=0.4, label='Inhabitant distr.', color='blue')

plt.yticks([y_pos[0]-0.4, y_pos[0], y_pos[1]-0.4, y_pos[1]],
           np.array(["Language requirement not met", "Non native", "Language requirement met", "Native"]))

plt.xlabel('Relative occurance (%)')
plt.title('Distribution of Dataset / inhabitants of Rotterdam')

plt.legend()

# Show the plot
plt.tight_layout()
plt.show()

```
