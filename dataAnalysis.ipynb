**Data Analysis**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Let's load the dataset
data = pd.read_csv('data/all_data.csv')

# Let's specify the features and the target
y = data['checked']
X = data.drop(['checked'], axis=1)
X = X.astype(np.float32)
```

```python
for i in (list(data.columns.values)):
    print(i)
```

```python
geslacht_vrouw = data['persoon_geslacht_vrouw']
counts = geslacht_vrouw.value_counts()
print(counts)

plt.bar(counts.index, counts.values, color=['blue', 'orange'])

# Add labels and title
plt.ylabel('Count')
plt.title('Amount of men and women')

plt.xticks([0, 1], labels=['men', 'women'])
plt.plot
```

```python
wijken = data.filter(like='adres_recentste_wijk', axis=1)
wijken_count = wijken.sum(axis=0)

percentages = (wijken_count / len(wijken)) * 100

# Plot the histogram (bar chart)
plt.bar(percentages.index, percentages.values, color='skyblue')

# Add labels and title
plt.xlabel('Columns')
plt.ylabel('Percentage of 1s (%)')
plt.title('Percentage of 1s for each \'wijk\'')

# Show the plot
plt.show()
```
